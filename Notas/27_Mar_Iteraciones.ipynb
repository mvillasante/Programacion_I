### Ejercicio: Suma de Gauss

La suma de los primers *n* números naturales la calculó Gauss a muy temprana edad:

$S(n)= \frac{n(n+1)}{2}$

Nosotros vamos a utilizar un `for` para calcular la suma y comparar que el resulta es el adecuado.

````julia
n=1000
suma = 0
for i in 1:n
    println("Iteracion: ", i)
    println(suma)
    suma = i + suma
end
println(suma)
````

````julia
function suma_gauss(n)
    return n*(n+1)/2
end

suma_gauss(n) == suma
````

En **Julia** los `for` pueden iterar sobre distintos objetos, además de los rangos:

* Sobre **cadenas**

````julia
for x in "Aceptan como rango casi cualquier cosa"
    @show x
end
````

* Sobre **arreglos**

````julia
vector = [1,3,6,"hola"]
````

````julia
for x in vector
    @show x
end
````

````julia
function gauss_mano(n:: Int64)
    if !(n isa Int64) || (n<0)
        error("La fórmula de Gauss tiene sentido para los naturales")
    else
        suma = 0
        for i in 1:n
            suma += i 
        end
    return suma
    end
end
````

````julia
gauss_mano(100)
````

````julia
gauss_mano(-10)
````

````julia
gauss_mano("hola")
````

## `while`

Este método repetirá las declaraciones dentro de su estructura hasta que deje de ser verdadera la condición que indiquemos. La sentencia `while` tiene la siguiente estructura:

```julia
while <condición>
    <declaraciones>
end
```
Esta estructua se puede leer como:

> "Mientas la condición sea verdadera, la computadora ejecutará las declaraciones dentro del `while`"

El flujo de ejecución de este método es el siguiente:
1. Se determina si la condición es verdadera.
2. Si es falsa, la computadora ignora todas las declaraciones dentro del `while` y continua el flujo de ejecución después del `end`.
3. Si es verdadera ejecuta las declaraciones dentro del cuerpo del `while` y al terminar regresa al paso 1.



````julia
i=5
while i <=10
    @show i<10
    i += 1
end
````

## Calculando raices 

Dentro de las tantas contribuciones que Newton hizo a las matemáticas está el siguiente algoritmo para calcular raíces.

Suponemos que queremos calcular la raíz cuadrada de `a`.

$ y = \frac{1}{2}(x+a/x)$,

donde `x` es una valor aproximado a la raíz.

Supongamos que queremos calcular la raíz cuadrada de 4, es decir `a=4`, usando como aproximación inicial `x=3`

````julia
a = 4
x = 3

y = (x+a/x)/2
````

El valor de `y` se acerca al valor exacto. Ahora vamos a tomar este valor de `y` como la nueva aproximación en el método de Newton:

````julia
x = y
y = (x+a/x)/2
````

````julia
x = y
y = (x+a/x)/2
````

````julia
x = y
y = (x+a/x)/2
````

Al hacer este procedimiento solamente tres veces ya tenemos una aproximación bastante buena al valor exacto.
